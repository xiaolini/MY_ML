### matplotlib基础

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
cosy
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle='--')
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle='--')
plt.xlim(-1, 11)
plt.ylim(-2, 2)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle='--')
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red', linestyle='--')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, color = 'red', linestyle='--', label='cos(x)')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, color = 'red', linestyle='--', label='cos(x)')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("title")
plt.show()
```

### 散点图

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color='red')
plt.show()
```
